So lets extract all the adjectives from the text <br>
and look at the vectors

```python
adj_dict = dict([ (tt.lower_, tt) for tt in tokens
                         if len(tt.orth_) > 5 and  tt.pos_  in ['ADJ']])

adj_tokens = [adj_dict[x] for x in sorted(adj_dict.keys())]
```

Normalise the vectors because
I'm going to use normalised dot product as a distance

```python
vecs  = [(x, x.vector/x.vector_norm ) for x in adj_tokens
                                       if x.has_vector]
len(vecs)
```

```python
import numpy as np
```

we want to do a calculation like this **example**

```python
M = np.random.rand(5,5)
X = np.dot(M,M.transpose())
np.linalg.eig(X)
```

Calculate the dot products - 
this is a symmetric matrix so 
- diagonalisable
- eigenvectors are orthogonal

we are going to project onto 
the space spanned by the top two eigenvectors.

```python
MM = np.asarray( [y for x,y in vecs])
XX =  np.dot(MM,MM.transpose())
MM.shape, XX.shape
```

```python
evs = np.linalg.eig(XX)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
labels = [x[0] for x in vecs]
```

```python
plt.rcParams["figure.figsize"] = (10,10)

XS = evs[1][0]/np.linalg.norm(evs[1][0])
YS = evs[1][1]/np.linalg.norm(evs[1][1])

plt.plot(XS,YS,'or')
for label, x, y in zip(labels[:],XS,YS):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-10, 10),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))
```
